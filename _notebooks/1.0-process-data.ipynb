### Imports

```python
import pandas as pd
import numpy as np
import capir_transfronteriza2_2023.data.load as load
import capir_transfronteriza2_2023.data.process as process
```

### Load data

```python
data_minet = load.data_minet
data_processed = load.data_processed
```

### Read data

```python
files = process.get_csv_files(data_minet)
```

```python
dfs = []
for f in files:
    df = pd.read_csv(f, low_memory=False)
    dfs.append(df)
```

```python
df = pd.concat(dfs)

print(df.shape)

df.head(3)
```

```python
# Adding column for country
conditions = [
    (df['user_screen_name'] == "nikolas_dm") | (df['user_screen_name'] == "PastorMalafaia") | (df['user_screen_name'] == "brasilsemaborto"),
    (df['user_screen_name'] == "MamelaFialloFlo") | (df['user_screen_name'] == "etorrescobo") | (df['user_screen_name'] == "_FamiliaEcuador"),
    (df['user_screen_name'] == "misionpaz_") | (df['user_screen_name'] == "MariaFdaCabal") | (df['user_screen_name'] == "UnidosxlaVidaCo"),
    (df['user_screen_name'] == "RoySantosC") | (df['user_screen_name'] == "TommyZambranoM") | (df['user_screen_name'] == "luisffaraj"),
    ]

values = ['Brasil', 'Ecuador', 'Colombia', 'Honduras']

df['country'] = np.select(conditions, values)

df.head(3)
```

```python
df.to_csv(f'{data_processed}/tweets.csv', index=False)
```
