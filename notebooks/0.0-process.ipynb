```python
import re
import pandas as pd

df = pd.read_csv("../data/raw/brasilsemaborto.csv", header=None, names=['url'])
df
```

```python
df['id'] = df['url'].apply(lambda x: re.findall(r'twitter.com\/\w+\/status\/(\d+)', str(x))[0])
df
```

```python
ids = df['id'].to_list()
with open(r'../data/txt/ids_brasilsemaborto.txt', 'w') as fp:
    fp.write('\n'.join(ids))
```
